```python
from scipy.stats import ttest_ind, chi2_contingency

# Example: T-test for sales in different cyties

# Assuming 'total_sales' is the column with sales data
# and 'City' is the column with city information.
# Replace 'City A' and 'City B' with actual city names.

city_a_sales = df[df['City'] == 'City A']['total_sales']
city_b_sales = df[df['City'] == 'City B']['total_sales']

# Perform T-test assuming equal variances
t_stat, p_value = stats.ttest_ind(city_a_sales, city_b_sales)
print(f"T-test result: t-statistic = {t_stat}, p-value = {p_value}")
```

```python
#Selecting only numeric columns for correlation matrix

import pandas as pd

df=pd.read_csv('/content/Sales Data.csv')
# Check the data types of all columns
print(df.dtypes)

# Select only numeric columns from the DataFrame
numeric_df = df.select_dtypes(include=['number'])
```

```python
# Compute the correlation matrix
correlation_matrix = numeric_df.corr()

# Print the correlation matrix
print(correlation_matrix)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot the correlation matrix as a heatmap
plt.figure(figsize=(10, 8))  # Adjust size as needed
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
plt.title("Correlation Matrix")
plt.show()
```
